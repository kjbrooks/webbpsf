# Constructing an OPD after maximum thermal slew

This notebook will walk through creating a JWST OPD for a given time after the observatory has undergone a *maximum* slew relative to the sun. 



This work is based off of five thermal OPD models we have for 10,000 seconds, 1 day, 7 days, 14 days, and 30 days. We fit these data with a second order thermal response function and saved the fitting parameters to a FITS table, which is what is used to create these final OPDs. 

```python
#stdlib
import os
from collections import OrderedDict

#Third Party
import datetime
from astropy.io import fits
from astropy.table import Table
import astropy.units as u
import numpy as np
import poppy
from scipy.io import loadmat
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import matplotlib
import webbpsf
from webbpsf import constants
plt.rcParams.update({'figure.dpi' : 100,
                     'font.size' : 12,
                     'image.origin' : 'lower',
                     'image.interpolation' : 'nearest'})

%matplotlib inline
```

# Construct the surfaces at different times after maximum thermal slew 

## $$\Delta t = 15 mins $$

```python
delta_time = 15 * u.min # Delta time should be an astropy units quantity
otelm = webbpsf.opds.OTE_Linear_Model_WSS()
otelm.thermal_slew(delta_time=delta_time)

plt.figure(figsize=(10,8))
plt.imshow(otelm.opd)
plt.title("OPD for delta time = {}".format(delta_time))
cbar = plt.colorbar()
cbar.set_label('m')
plt.show()
```

# Construct the surfaces at different times after a slew of some angle

## $$\Delta t = 15 mins $$

```python
delta_time = 15 * u.min # Delta time should be an astropy units quantity
otelm = webbpsf.opds.OTE_Linear_Model_WSS()
otelm.thermal_slew(delta_time=delta_time, start_angle=10., end_angle=35.)

plt.figure(figsize=(10,8))
plt.imshow(otelm.opd)
plt.title("OPD for delta time = {}".format(delta_time))
cbar = plt.colorbar()
cbar.set_label('m')
plt.show()
```
